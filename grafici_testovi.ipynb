```python
import matplotlib.pyplot as plt

# Define the time values
t = [0, 2, 3, 4]
# Define the position values corresponding to the time points
x = [-10, 30, 20, 20]

# Create the plot
plt.plot(t, x, marker='o', linestyle='-', color='b')
plt.xlabel('t (s)')
plt.ylabel('x (m)')

# Make the axes more prominent with arrows at the end
plt.axhline(0, color='black', linewidth=0.5, linestyle='-', alpha=0.7)
plt.axvline(0, color='black', linewidth=0.5, linestyle='-', alpha=0.7)

# Remove the box around the figure
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# Make the grid less prominent
plt.grid(True, alpha=0.2)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
font = 16

# Define the time values
t = [0, 0.5, 2, 3, 4]
# Define the position values corresponding to the time points
x = [-10, 0, 30, 20, 20]

# Create the plot with improved quality and font size
plt.figure(dpi=150)
plt.plot(t, x, marker='o', linestyle='-', color='b', markersize=4)
plt.xlabel('t (s)', fontsize=font)
plt.ylabel('x (m)', fontsize=font)

# Make the axes more prominent with arrows at the top and right
plt.axhline(0, color='black', linewidth=0.5, linestyle='-', alpha=0.7)
plt.axvline(0, color='black', linewidth=0.5, linestyle='-', alpha=0.7)

# Remove the box around the figure
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# Show the plot with less prominent grid lines
plt.grid(True, alpha=0.2)

# Increase the quality of the figure
plt.tight_layout()
plt.show()
```
